```
 # Loading packages requiresd for the Feed Forward Neural 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from keras.utils import to_categorical
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import classification_report
```

```
# Loading the training and test Dataset
train_dataset = pd.read_csv('train.csv')
test_dataset = pd.read_csv('test.csv')
```

```
train_dataset['Activity'] = train_dataset['Activity'].map({"STANDING": 0, "SITTING": 1, "LAYING": 2, "WALKING": 3, "WALKING_DOWNSTAIRS": 4, "WALKING_UPSTAIRS": 5})
```

```
test_dataset['Activity'] = test_dataset['Activity'].map({"STANDING": 0, "SITTING": 1, "LAYING": 2, "WALKING": 3, "WALKING_DOWNSTAIRS": 4, "WALKING_UPSTAIRS": 5})
```

```
features_cols = set(train_dataset.columns)
features_cols.remove('Activity')
```

```
print(features_cols)
```

```
Label_cols = ['Activity']
```

```
# Training Set
X_train = train_dataset.loc[:, features_cols].values
y_train = train_dataset.loc[:, Label_cols].values
```

```
X_train
```

```
y_train[9:40]
        
```

```
# Convert y_train digits into one hot coding
y_train = to_categorical(y_train)
```

```
y_train[9:40]
```

```
# Test Set
X_test = test_dataset.loc[:, features_cols].values
y_test = test_dataset.loc[:, Label_cols].values
```

```
y_test_onehot = to_categorical(y_test)
```

```
# Create Feedforward model
# Fit the model for the X_train and y_train(one hot coding values)
# Predict teh Y_values from the X_test data
# Initializing the Feedforward Neural Network
model  = Sequential()
# Designing the first hidden layer
model.add(Dense(12, input_dim = X_train.shape[1], activation='sigmoid'))
# Designing the second hidden layer
#model.add(Dense(8, activation='sigmoid'))
# Designing the output layer
model.add(Dense(y_train.shape[1], activation='softmax'))
# Compile the network with the loss function is loss = 'categorical_crossentropy' and optimizer is adam(will update weights)
# metrics model is looking w.r.t accuracy
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
# Fit the model with training and validation data sets
history = model.fit(X_train, y_train,
                    batch_size = 32, epochs = 20,
                    validation_data = (X_test, y_test_onehot))
# Model summary for number of parameters use in the algorithm
model.summary()
```

```
_, accuracy = model.evaluate(X_test, y_test_onehot)
print('Accuracy: %.2f' %(accuracy*100))
```

```
y_pred  = model.predict(X_test, batch_size = 64, verbose = 1)
# Converting the one hot coding result to picking the index of max class
y_pred_bool = np.argmax(y_pred, axis = 1)
# Classification result of each class
print(classification_report(y_test, y_pred_bool))
```
